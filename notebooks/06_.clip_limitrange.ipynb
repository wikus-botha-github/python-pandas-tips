#### Data Clipping using .clip()

`.clip()` caps the values in a column to a specified minimum and/or maximum — a quick way to tame outliers without dropping any rows.

```python
import pandas as pd
```

```python
data = {
    "Date": pd.date_range(start="2023-11-12", periods=10, freq="D"),
    "Temperature": [6, 19, 28, 14, 10, 7, 28, 20, 6, 25]
}

df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
```

Clip the `Temperature` column so that any value below 10 becomes 10, and any value above 20 becomes 20.

```python
df["Temperature_Clipped"] = df["Temperature"].clip(lower=10, upper=20)
```

```python
display(df)
```

A few things to keep in mind:
- `lower` sets the floor — any value below it is replaced with the floor value.
- `upper` sets the ceiling — any value above it is replaced with the ceiling value.
- Values already inside the range are left untouched, unlike binning which reassigns every value to a category.
