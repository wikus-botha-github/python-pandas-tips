#### Separating Weekdays and Weekends

When a DataFrame has a `DatetimeIndex`, the `.weekday` attribute returns an integer for each date (Monday = 0 ... Sunday = 6). This makes it simple to split a time series into weekday and weekend subsets.

```python
import pandas as pd
```

```python
data = {
    "Date": pd.date_range(start="2023-12-01", periods=7, freq="D"),
    "Temperature": [16, 13, 27, 17, 14, 16, 19]
}

dummy_data = pd.DataFrame(data)
dummy_data.set_index("Date", inplace=True)
```

Use a boolean mask on `index.weekday` to separate the two groups — weekdays are 0-4, weekends are 5-6.

```python
weekdays = dummy_data[dummy_data.index.weekday < 5]
weekends = dummy_data[dummy_data.index.weekday >= 5]
```

```python
display(weekdays)
```

```python
display(weekends)
```

A few things to keep in mind:
- `.weekday` returns `0` for Monday through `6` for Sunday.
- The threshold `< 5` captures weekdays, while `>= 5` captures Saturday and Sunday.
- `.dayofweek` is an identical alias for `.weekday` if you prefer that naming.
