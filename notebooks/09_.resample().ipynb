#### Data Resampling using .resample()

`.resample()` changes the frequency of time series data — for example, converting hourly readings into a daily summary — and pairs naturally with an aggregation like `.max()` or `.mean()`.

```python
import pandas as pd
import numpy as np

np.random.seed(1)
```

```python
dummy_data = pd.DataFrame({
    "Temperature": np.random.randint(5, 30, size=72)
}, index=pd.date_range("2023-12-09", periods=72, freq="h"))
```

Resample the hourly data down to a daily frequency and take the maximum temperature recorded on each day.

```python
daily_temp_max = dummy_data.resample("D")["Temperature"].max()
```

```python
display(daily_temp_max)
```

A few things to keep in mind:
- `'D'` is a daily frequency; other common aliases include `'H'` (hourly), `'W'` (weekly) and `'M'` (month end).
- `.resample()` behaves much like `.groupby()`, except it groups by time buckets instead of column values.
- Any aggregation function (`.max()`, `.mean()`, `.sum()`, ...) can follow the resample call.
