#### Data Binning using pd.cut()

`pd.cut()` lets you bin continuous numerical data into discrete intervals or categories — handy for turning a raw measurement like temperature into a simple label such as 'Cold' or 'Warm'.

```python
import pandas as pd
```

```python
data = {
    "Date": pd.date_range(start="2023-11-07", periods=10, freq="D"),
    "Temperature": [16, 13, 22, 24, 20, 17, 22, 14, 16, 19]
}

df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
```

Bin the `Temperature` column into two categories, 'Cold' and 'Warm', using the bin edges `[0, 16, 30]`.

```python
df["Temperature_Bin"] = pd.cut(
    df["Temperature"],
    bins=[0, 16, 30],
    labels=["Cold", "Warm"]
)
```

```python
display(df)
```

A few things to keep in mind:
- `bins` defines the edges of each interval — `[0, 16, 30]` creates two bins: `(0, 16]` and `(16, 30]`.
- `labels` names each bin in the same order as the intervals.
- By default, `pd.cut()` bins are right-closed, meaning a value equal to the upper edge is included in that bin.
