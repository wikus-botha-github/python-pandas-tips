#### Looking into Lead and Lag using .shift()

The `.shift()` method moves the values in a column up or down by a given number of periods. This makes it easy to compare a row to the previous row (lag) or the next row (lead) - for example, comparing today's temperature to yesterday's or tomorrow's.

```python
import pandas as pd
```

```python
data = {
    "Date": pd.date_range(start="2023-11-05", periods=10, freq="D"),
    "Temperature": [16, 29, 24, 20, 7, 16, 28, 20, 20, 13]
}

df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
```

Shift the `Temperature` column by 1 period to get the **previous** day's value (lag), and by -1 to get the **next** day's value (lead).

```python
df["Temperature_PrevDay"] = df["Temperature"].shift(periods=1)
df["Temperature_NextDay"] = df["Temperature"].shift(periods=-1)
```

```python
display(df)
```

We can also use the lagged column to calculate the day-over-day change in temperature.

```python
df["Temperature_DailyChange"] = df["Temperature"] - df["Temperature_PrevDay"]
```

```python
display(df)
```

A few things to keep in mind:
- `shift(periods=1)` moves values **down**, giving you the previous row's value.
- `shift(periods=-1)` moves values **up**, giving you the next row's value.
- The first (or last) row(s) will contain `NaN` since there is no prior (or subsequent) row to shift from.
