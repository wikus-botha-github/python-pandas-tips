#### Time Series Segmentation using .between_time()

`.between_time()` selects rows whose *time of day* falls within a given start and end time, regardless of the date. It's a handy way to split a day into segments such as early morning, morning, afternoon and evening.

```python
import pandas as pd
import numpy as np

np.random.seed(0)
```

```python
dummy_data = pd.DataFrame({
    "Temperature": np.random.randint(5, 30, size=24)
}, index=pd.date_range("2023-11-28", periods=24, freq="h"))
```

Split the day into four time-of-day segments.

```python
early_morning_data = dummy_data.between_time("00:00", "06:00")
morning_data = dummy_data.between_time("06:00", "12:00")
afternoon_data = dummy_data.between_time("12:00", "18:00")
evening_data = dummy_data.between_time("18:00", "23:59")
```

```python
display(early_morning_data)
display(morning_data)
display(afternoon_data)
display(evening_data)
```

A few things to keep in mind:
- `.between_time()` requires the DataFrame to have a `DatetimeIndex`.
- By default both the start and end time are inclusive (controlled by the `inclusive` parameter).
- Because it matches on time of day only, it works seamlessly across multiple days in the same call.
