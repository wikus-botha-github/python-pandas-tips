#### Missing Data Imputation (Backward Fill)

Backward fill (`bfill`) fills a missing value using the **next** valid observation in the column — useful when it makes more sense to assume a gap took on the value that followed it.

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Date": pd.date_range(start="2023-01-13", periods=7, freq="D"),
    "Temperature": [20.0, 18.0, np.nan, 22.0, 21.0, np.nan, 23.0]
}

dummy_data = pd.DataFrame(data)
dummy_data.set_index("Date", inplace=True)
```

Fill the missing values by looking forward to the next valid reading.

```python
dummy_data["bfill_Temperature"] = dummy_data["Temperature"].bfill()
```

```python
display(dummy_data)
```

A few things to keep in mind:
- Backward fill copies the **next** non-null value backward into each gap.
- If the last row(s) are missing, there is nothing ahead to fill from, so they remain `NaN`.
- `.bfill()` is the modern, preferred syntax — older code may use the now-deprecated `fillna(method="bfill")`, which does the same thing.
