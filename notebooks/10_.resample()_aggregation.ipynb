#### Data Resampling with Aggregation

Combining `.resample()` with `.agg()` lets you apply a *different* aggregation function to each column in a single call.

```python
import pandas as pd
import numpy as np

np.random.seed(2)
```

```python
dummy_data = pd.DataFrame({
    "Temperature": np.random.randint(5, 30, size=72),
    "Humidity": np.random.uniform(40, 80, size=72),
    "Wind Speed": np.random.uniform(0, 15, size=72)
}, index=pd.date_range("2023-12-11", periods=72, freq="h"))
```

Resample to a daily frequency, taking the daily **max** temperature alongside the daily **mean** humidity and wind speed.

```python
daily_aggregates = dummy_data.resample("D").agg({
    "Temperature": "max",
    "Humidity": "mean",
    "Wind Speed": "mean"
})
```

```python
display(daily_aggregates)
```

A few things to keep in mind:
- The dictionary passed to `.agg()` maps each column name to the aggregation function that should be applied to it.
- You can mix built-in strings (`'max'`, `'mean'`, `'sum'`) with custom functions.
- This is far more efficient than resampling each column separately.
